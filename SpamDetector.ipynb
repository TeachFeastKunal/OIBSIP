```python
import kagglehub
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import userdata
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report , mean_absolute_error , mean_squared_error , r2_score , accuracy_score
```

```python
os.environ['KAGGLE_USERNAME'] = userdata.get('KAGGLE_USERNAME')
os.environ['KAGGLE_KEY'] = userdata.get('KAGGLE_KEY')
```

```python
csv_path = kagglehub.dataset_download('uciml/sms-spam-collection-dataset')
file_path = os.listdir(csv_path)[0]
```

```python
df = pd.read_csv(os.path.join(csv_path , file_path) , encoding = 'latin-1')
df = df.dropna(axis =1)
df.columns = ['Category' , 'Message']

df['Category'] = df['Category'].map({'ham' : 0 , 'spam' : 1})
```

## Modelling

```python
x_train , x_test , y_train , y_test = train_test_split(df['Message'] , df['Category']  , test_size = 0.3 , random_state = 10)

# vectorizing English words

vectorize = TfidfVectorizer(stop_words = 'english')
x_vec_train = vectorize.fit_transform(x_train)
x_vec_test = vectorize.transform(x_test)
```

```python
# Train Bayes Model

model = MultinomialNB()
model.fit(x_vec_train , y_train)

y_pred = model.predict(x_vec_test)
```

## Model Evaluation

```python
classification_rep = classification_report(y_test , y_pred)
r2score = r2_score(y_test , y_pred)
acc_score = accuracy_score(y_test , y_pred)

print('classification report :\n' , classification_rep)
print('r2 score' , r2score)
print('accuracy score' , acc_score)
```
